```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')

print(df.shape)
print(df.columns)
display(df.head())

df.info()
display(df.describe())
```

```python
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.show()
```

```python
df.hist(figsize=(10, 8))
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.show()
```
